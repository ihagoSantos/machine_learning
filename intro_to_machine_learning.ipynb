# Importação das Bibliotecas

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# Carregamento da Base

```python
iris = load_iris()
X = iris['data']
y = iris['target']
```

# Exploração de dados

```python
print("Campos to dataset Iris ({0}): \n{1}\n".format("iris.keys()", iris.keys()))
print("Parte da descrição do dataset Iris ({0}): \n{1}\n".format("iris['DESCR'][:500]", iris['DESCR'][:500]))
print("Nomes das classes ({0}): {1}".format("iris['target_names']", iris['target_names']))
print("Nomes das features ({0}): {1}".format("iris['feature_names']", iris['feature_names']))
```

# Divisão da base original em bases e treino e teste

```python
# Por padrão, a função train_test_slipt particiona a base reservando 75% dos exemplos para treinamento e 25% para testes. Mas esses valores podem ser especificado na passagem da função.

X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0)
print("Base de treinamento ({0}): {1}".format("X_train.shape", X_train.shape))
print("Base de teste ({0}): {1}".format("X_test.shape", X_test.shape))
```
